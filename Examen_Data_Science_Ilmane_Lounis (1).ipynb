## 5 - Analyse multi-variée


Intéressons nous à la reussite ou non d'un projet.
Un projet est réussit si le montant du funding (pledge) est égal ou supérieur montant du funding attendu(goal).
Verifions cela en corrélant ces deux montants et la réussite ou non du projet :

```python
g = sns.PairGrid(df, vars=["pledged", "goal"], hue='state', palette='RdBu_r')
g.map(mp.scatter, alpha=1)
g.add_legend();
```

Bien que ces deux montants determinent la réussite ou non du projet, la corrélation entre ce deux derniers n'est pas vriament significative.

```python
df.loc[:, ['goal', 'pledged']].corr()
```

Sachant qu'un projet est réussi, quelles vont donc être la ou les variables qui vont influer sur le funding (pledge) ?


Interressons nous au financement avecune heatmap en fonction des pledges, goals et des backers :

```python
sns.heatmap(df.corr(),annot=True)
```

On peut s'apercevoir qu'il y a une forte corrélation entre les backers et les goals, pledge.
On peut poser l'hypothèse que la réussite du projet va dépendre du nombre de backers.



Interessons nous maintenant aux catégories.

Est-ce que la catégorie va influer sur le nombre de backers, et donc sur la réussite ou non du projet ?

## Conclusion

Sachant qu'un projet est réussi lorsque le funding est égal ou supérieur au montant attendu (goal). L'hypothèse concernant le nombre de backers semble fiable, cela signifie que pour tout projet, il n'existe pas d'unique donateurs et que donc le funding à tendance à etre reparti sur beaucoup plus de donateurs.

Il faut cependant regarder ce qui va influer sur le nombre de backers, car on ne peut se limiter à cette hypothèse.
